```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
from numpy.random import randn,randint,uniform,sample
```

```python
x =np.arange(0,10)
y= x*x

plt.plot(x,y,'r^',linestyle='dashed', markersize=15)
plt.title('Square of x Chart')
plt.xlabel('x values',color='blue')
plt.ylabel('square of x',color='blue')
```

```python
x=[1,2,3,4,5,6,7]   #days
y=[160,150,140,145,175,165,180] #sales

plt.title('Ass Day 1')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y,'bo',linestyle='dashed', markersize=10)
```

```python
days=[1,2,3,4,5,6,7]   #days
sales1=[160,150,140,145,175,165,180] 
sales2=[70,90,160,150,140,145,175] 

plt.figure(figsize=(6,5),dpi=100)
plt.plot(days,sales1,color='blue')
plt.plot(days,sales2,color='red')


plt.xlabel("Days of the week")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

```python
#Create a 3 by 3 subplots:

x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]


plt.subplot(3,3,1)
plt.plot(x,y1,'r--')
plt.subplot(3,3,2)
plt.plot(x,y2,'g*--')
plt.subplot(3,3,3)
plt.plot(x,y3,'b--')
plt.subplot(3,3,4)
plt.plot(x,y4,'g--')
plt.subplot(3,3,5)
plt.plot(x,y4,'go')
```
